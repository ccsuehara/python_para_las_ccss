```python
import requests #html requestor
from bs4 import BeautifulSoup #html parser
import pandas as pd #dataframe manipulator
```

```python
## Vamos a abrir una página web y extraer la información de la tabla de la página.
#  Esta versión del catálogo de cursos es anterior, pero nos ayuda a entender la lógica del webscrapping. 
#https://stackoverflow.com/questions/21570780/using-python-and-beautifulsoup-saved-webpage-source-codes-into-a-local-file

with open('./data/Listado de cursos - Facultad de Ciencias Sociales.html',
          'r', encoding='utf-8') as file:
    soup = BeautifulSoup(file, 'html.parser')
```

```python
#print(soup.prettify())
```

```python
cursos = soup.find_all('h2')
cursos[0].get_text()
```

```python
descripcion = soup.find_all('p')


descripcion[1].get_text()
```

```python
cursos2 = soup.find_all('div', {"class":["col-sm-12 listas-internas"]})
```

```python
nombres = []
for curso in cursos:
    nombres.append(curso.text)
```

```python
descripciones = []
nombres = []
creditos = []
for descrip in cursos2:
    nombre = descrip.find('h2').text
    des = descrip.find('p').text
    cred = descrip.find('span').text
    
    creditos.append(cred)
    descripciones.append(des)
    nombres.append(nombre)
    #print(nombres)
```

```python
df = pd.DataFrame()
df['nombre'] = nombres
df['descripcion'] = descripciones
df['credito'] = creditos
```
